### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
type(tienda)
```

```python
Total = tienda["Precio"].sum()
Total2 = tienda2["Precio"].sum()
Total3 = tienda3["Precio"].sum()
Total4 = tienda4["Precio"].sum()
print(f'Total facturado tienda 1: $ {Total}')
print(f'Total facturado tienda 2: $ {Total2}')
print(f'Total facturado tienda 3: $ {Total3}')
print(f'Total facturado tienda 4: $ {Total4}')
Total_final = Total + Total2 + Total3 + Total4
print(f'Total geeneral por las 4 tiendas: ${Total_final}')
```

# 2. Ventas por categoría

```python
# ventas por categoria por tienda
ventas_por_categoria_tienda1 = tienda.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
ventas_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
ventas_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
ventas_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)

print("Ventas por categoría en Tienda 1:")
print(ventas_por_categoria_tienda1)
print("\nVentas por categoría en Tienda 2:")
print(ventas_por_categoria_tienda2)
print("\nVentas por categoría en Tienda 3:")
print(ventas_por_categoria_tienda3)
print("\nVentas por categoría en Tienda 4:")
print(ventas_por_categoria_tienda4)

# cantidad de productos vendidos por categoria
productos_por_categoria_tienda1 = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)
productos_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False)
productos_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)
productos_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False)

print("\nNúmero de productos vendidos por categoría en Tienda 1:")
print(productos_por_categoria_tienda1)
print("\nNúmero de productos vendidos por categoría en Tienda 2:")
print(productos_por_categoria_tienda2)
print("\nNúmero de productos vendidos por categoría en Tienda 3:")
print(productos_por_categoria_tienda3)
print("\nNúmero de productos vendidos por categoría en Tienda 4:")
productos_por_categoria_tienda4
```

## Combinar dataframes

### Subtask:
Unir los DataFrames de las cuatro tiendas en un único DataFrame.


**Reasoning**:
Concatenate the four dataframes into a single dataframe and display the head of the new dataframe to verify the concatenation.



```python
import pandas as pd

# Unir los cuatro DataFrames en uno solo
combined_df = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

# Guardar el DataFrame combinado en un nuevo archivo CSV
#combined_df.to_csv('tiendas_combinadas.csv', index=False)

#print("Los cuatro DataFrames se han unido y el resultado se ha guardado en 'tiendas_combinadas.csv'")
```

```python
# Agregar una columna 'Tienda' a cada DataFrame individual
tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Unir los cuatro DataFrames en uno solo con la nueva columna
combined_df_with_source = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

# Mostrar las primeras filas del nuevo DataFrame combinado para verificar
print("DataFrame combinado con la columna 'Tienda':")
display(combined_df_with_source.head())
```

```python
import matplotlib.pyplot as plt

# Calcular la facturación total (Precio + Costo de envío) en el DataFrame combinado
combined_df_with_source['Facturacion_Total'] = combined_df_with_source['Precio'] + combined_df_with_source['Costo de envío']

# Agrupar por tienda y sumar la facturación total
facturacion_por_tienda_combinado = combined_df_with_source.groupby('Tienda')['Facturacion_Total'].sum()

# Crear la gráfica de pastel
plt.figure(figsize=(8, 8))
plt.pie(facturacion_por_tienda_combinado, labels=facturacion_por_tienda_combinado.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de la Facturación Total por Tienda (Precio + Costo de Envío)')
plt.axis('equal')  # Asegura que el gráfico de pastel sea un círculo.
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrar el DataFrame combinado para incluir solo las calificaciones de 5 estrellas
calificaciones_5_estrellas = combined_df_with_source[combined_df_with_source['Calificación'] == 5]

# Contar cuántas calificaciones de 5 estrellas tiene cada tienda
conteo_5_estrellas_por_tienda = calificaciones_5_estrellas['Tienda'].value_counts().sort_index()

# Crear la gráfica de barras
plt.figure(figsize=(8, 6))
sns.barplot(x=conteo_5_estrellas_por_tienda.index, y=conteo_5_estrellas_por_tienda.values, palette='viridis')

plt.title('Cantidad de Calificaciones de 5 Estrellas por Tienda', fontsize=14)
plt.xlabel('Tienda')
plt.ylabel('Cantidad de Calificaciones de 5 Estrellas')
plt.ylim(0, conteo_5_estrellas_por_tienda.max() * 1.1) # Ajustar el límite del eje y para espacio de etiquetas

# Añadir etiquetas con los valores en las barras
for index, value in enumerate(conteo_5_estrellas_por_tienda.values):
    plt.text(index, value, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular el conteo de cada producto a nivel global
global_product_counts = combined_df_with_source['Producto'].value_counts()

# Seleccionar los 10 productos más vendidos a nivel global
top_10_global_products = global_product_counts.head(10).index.tolist()

# Filtrar el DataFrame combinado para incluir solo estos 10 productos
filtered_df = combined_df_with_source[combined_df_with_source['Producto'].isin(top_10_global_products)]

# Contar las ventas de estos productos por tienda
sales_by_product_and_store = filtered_df.groupby(['Producto', 'Tienda']).size().unstack(fill_value=0)

# Crear la gráfica de barras agrupadas
ax = sales_by_product_and_store.plot(kind='bar', figsize=(15, 8), width=0.8)

plt.title('Ventas de los 10 Productos Más Vendidos Globalmente por Tienda', fontsize=16)
plt.xlabel('Producto')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45, ha='right') # Rotar etiquetas del eje x para mejor lectura
plt.legend(title='Tienda', bbox_to_anchor=(1.05, 1), loc='upper left') # Mover la leyenda

plt.tight_layout()
plt.show()
```
